```python
import random

class Species():
    def __init__(self, n0, r):
        self.n = n0
        self.r = r
        
    def reproduce(self):
        self.n = self.n + self.r * self.n + random.normalvariate(0, 1)
        
    def get_n(self):
        return self.n
```

```python
populations = []
for i in range(100):
    populations.append(Species(10, 0.1))
```

```python
t = 0
while t < 50:
    for pop in populations:
        pop.reproduce()
    t +=1
```
